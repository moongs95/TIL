### 감탄 떡볶이의 입점전략은 파리바게트옆인가?
- 감탄떡볶이의 입점전략이 파리바게트 옆인지 알아보기위해
- 감탄떡볶이외 국대떡볶이, 죠스떡볶이, 엽기떡볶이, 신전떡볶이와 파리바게트의 거리를 구해서 비교

```python
pd.options.display.float_format = '{:.4f}'.format # 소수점 네자리 까지만 출력
url = 'https://raw.githubusercontent.com/panda-kim/csv_files/main/food_business_seoul_20221231.csv'
tt = ['감탄떡볶이', '국대떡볶이', '죠스떡볶이', '엽기떡볶이', '신전떡볶이']
```

```python
df = pd.read_csv(url)
df
```

```python
# 파리바게뜨 데이터 필터링하고 정제하기
cond1 = df['상호명'].str.contains(r'파리바게뜨|파리바게트')
df_paris = df.loc[cond1, ['상가업소번호', '경도', '위도']].copy()
df_paris = df_paris.set_axis(['p_업소번호', 'p_경도', 'p_위도'], axis=1).reset_index(drop=True)
df_paris
```

```python
# 떡볶이 프랜차이즈 데이터 필터링하고 정제하기
df_tt = df.copy()
# df_tt['상호명'].str.extract('(감탄떡볶이|국대떡볶이)').dropna()
df_tt['상호명'] = df_tt['상호명'].str.extract(r'({})'.format('|'.join(tt)))[0]
df_tt = df_tt.dropna(subset=['상호명']).iloc[:, [0, 1, 5, 8, 9]].reset_index(drop=True)
df_tt
```

```python
# 곱집합(cartesian product) & 위도와 경도로 거리구하기(haversine)
df1 = pd.DataFrame(['A', 'B'])
df2 = pd.DataFrame(['가', '나', '다'])
df1.merge(df2, how='cross') # cross join 모든 순서쌍 구함.
```

```python
# df_tt와 df_paris의 곱집합 만들기
df_cross = df_tt.merge(df_paris, how='cross')
df_cross
```

```python
# 두 가게 사이의 거리구하기
# haversine 라이브러리 설치하기 : 두 지점의 위도와 경도를 입력하면 거리를 구해줌
# !pip install haversine
```

```python
# haversine 사용법
from haversine import haversine
#  서울과 파리의 위도와 경도(위도 경도순)
seoul = [37.541, 126.986]
paris = [48.8567, 2.3508]

# 서울과 파리사이의 거리 계산
# haversine([37.541, 126.986], [48.8567, 2.3508], unit='m')
haversine(seoul, paris, unit='m') # 거리 m
```

```python
# apply로 haversine 함수 이용하여 거리 구하기
df_cross['거리'] = df_cross.apply(lambda x: haversine([x.iloc[4], x.iloc[3]], [x.iloc[7], x.iloc[6]], unit='m'), axis=1)
df_cross
```

```python
# 개별 떡볶이 매장과 파리바게뜨와의 거리
df_dis = df_cross.groupby(['상가업소번호', '상호명'])['거리'].min().reset_index()
df_dis
```

```python
# 프랜차이즈 별 파리바게뜨와의 평균 거리
df_dis.groupby('상호명')['거리'].mean()
```

```python
# 프랜차이즈 별 파리바게뜨와의 평균 거리와 매장개수
df_dis.groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# 프랜차이즈 별 파리바게뜨와 50m미만인 매장의 평균거리와 매장개수
cond = df_dis['거리'] < 50
df_dis[cond].groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# 프랜차이즈 별 파리바게뜨와 50m미만인 매장의 평균거리와 매장개수
cond = df_dis['거리'] < 30
df_dis[cond].groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# 거리에 따라 집계를 해주는 함수
def distance(x):
    cond = df_dis['거리'] < x
    return df_dis[cond].groupby('상호명')['거리'].agg(['mean', 'count'])

distance(100)
```

### 감탄떡볶이의 입점전략 시각화하기

```python
# 프로젝트 코드
import pandas as pd
from haversine import haversine
from pyecharts.charts import Timeline, Grid
from pandasecharts import echart
import IPython

pd.options.display.max_rows = 6 # 판다스 1.4+ 에서 6행까지만 출력
pd.options.display.float_format = '{:.4f}'.format # 소수점 네자리 까지만 출력
url = 'https://raw.githubusercontent.com/panda-kim/csv_files/main/food_business_seoul_20221231.csv'
tt = ['감탄떡볶이', '국대떡볶이', '죠스떡볶이', '엽기떡볶이', '신전떡볶이']
df = pd.read_csv(url)

# 파리바게뜨 데이터 필터링하고 정제하기
cond1 = df['상호명'].str.contains(r'파리바게뜨|파리바게트')
df_paris = df.loc[cond1, ['상가업소번호', '경도', '위도']].copy()
df_paris = (df_paris
            .set_axis(['p_업소번호', 'p_경도', 'p_위도'], axis=1)
            .reset_index(drop=True))

# 떡볶이 프랜차이즈 데이터 필터링하고 정제하기
df_tt = df.copy()
df_tt['상호명'] = df_tt['상호명'].str.extract(r'({})'.format('|'.join(tt)))[0]
df_tt = (df_tt
         .dropna(subset=['상호명'])
         .iloc[:, [0, 1, 5, 8, 9]]
         .reset_index(drop=True))

# df_tt와 df_paris의 곱집합 만들기
df_cross = df_tt.merge(df_paris, how='cross')

# df_cross에 거리열 만들기
df_cross['거리'] = df_cross.apply(
    lambda x: haversine([x.iloc[4],x.iloc[3]], [x.iloc[7], x.iloc[6]], unit='m'), 
    axis=1)

# 개별 떡볶이 매장과 파리바게뜨와의 거리
df_dis = df_cross.groupby(['상가업소번호', '상호명'])['거리'].min().reset_index()


# 거리에 따라 프랜차이즈별 평균 거리와 매장수를 집계를 해주는 함수
def distance(x):
  cond = df_dis['거리'] < x
  return df_dis[cond].groupby('상호명')['거리'].agg(['mean', 'count'])
```

```python
# distance 함수로 50m의 집계하기(reset_index로 상호명도 열로 만든다)
df_g = distance(50).reset_index()
df_g
```

```python
# 50m의 데이터로 원형 그래프 만들기
df_g.echart.pie(x='상호명', y='count', figsize=(600, 400),
               radius=['20%', '60%'], label_opts={'position':'outer'},
               title='감탄떡볶이의 입점전략은 과연 파리바게뜨 옆인가?',
               legend_opts={'pos_right':'0%', 'orient':'vertical'},
               subtitle='50m 이내 매장수').render()
IPython.display.HTML(filename='250829_pie_chart.html')
```

```python
# 앞서 만든 그래프를 timeline으로 만들기
tl = Timeline({'width':'600px', 'height':'400px'})
for i in [10000, 100, 50, 30]:
    df_g = distance(i).reset_index()
    pie1 = df_g.echart.pie(x='상호명',
                           y='count',
                           figsize=(600, 400),
                           radius=['20%', '60%'],
                           label_opts={'position':'outer'},
                           title='감탄떡볶이의 입점전략은 과연 파리바게뜨 옆인가?',
                           legend_opts={'pos_right':'0%', 'orient':'vertical'},
                           subtitle=f'{i}m 이내 매장수')
    tl.add(pie1, f'{i}m').render()
IPython.display.HTML(filename='250829_pie_timeline_chart.html')
```
